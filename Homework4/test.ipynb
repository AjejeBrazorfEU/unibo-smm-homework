```python
import numpy as np
import matplotlib.pyplot as plt

m = np.array([2,3])
def f(x):
    N = len(x)
    return m@x.T + 2*np.random.normal(0,1,(N,))

x1 = np.linspace(0,10).reshape((50,1))
x2 = np.linspace(0,10).reshape((50,1))
x = np.concatenate([x1,x2], axis=1)
print(x1.shape,x2.shape,x.shape)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x1, x2, f(x), c='r', marker='o')
```

```python
# Training data
x1 = np.linspace(0,10).reshape((50,1))
x2 = np.linspace(0,10).reshape((50,1))
X = np.concatenate([x1,x2], axis=1)
Y = f(X)
```

```python
print(X.T @ X)
estimation = np.linalg.inv(X.T @ X) @ X.T @ Y
```

```python
plt.plot(X, estimation*X)
plt.plot(X, Y, 'o')
print(estimation)
```

```python
def f(x):
    N = len(x)
    return 3*np.sin(x) + np.random.normal(0,1,(N,))
X = np.linspace(0,10)
Y = f(X)
plt.plot(X,Y,'o')
```

```python
def f(X, params):
    return np.flip(np.vander(X,k), 1) @ params
    
for k in [1,2,3,4,5,6,6,8,9,10]:
    PSI = np.flip(np.vander(X,k), 1)
    print(PSI.shape)

    parameters = np.linalg.inv(PSI.T @ PSI) @ PSI.T @ Y
    print(parameters.shape, parameters)

    plt.plot(X, f(X, parameters))
    plt.plot(X,Y,'o')
    plt.title(f'k = {k}')
    plt.show()
```
